I have 4 items to purchase.
These items are: apple, mango, carrot, banana, 
I also have to buy rice.
My shopping list is now: apple mango carrot banana rice 
I will sort my list now 
Sorted shopping list is: apple banana carrot mango rice 
The first item I will buy is: 
I bought the apple 
My shopping list is now: banana carrot mango rice 

```python
shopping_list=["apple", "mango", "carrot", "banana"]
print (f"I have {len(shopping_list)} items to purchase.")


print ("These items are:", end=" ")
for i in shopping_list:
    print(i, end=", ")
print ("\nI also have to buy rice.")


shopping_list.append("rice")
print("My shopping list is now:", end=" ")
for i in shopping_list:
    print(i, end=" ")


print("\nI will sort my list now", end=" ")
print("\nSorted shopping list is:", end=" ")
shopping_list.sort()
for i in shopping_list:
    print(i, end=" ")


print("\nThe first item I will buy is:", end=" ")
bought_item=shopping_list.pop(0)
print("\nI bought the",bought_item, end=" ")


print("\nMy shopping list is now:", end=" ")
for i in shopping_list:
    print(i, end= " ")
```

```python

import os
import zipfile
from datetime import datetime

backup_list = [
    "/Users/macbookpro/Desktop/English/lala.png",
    "/Users/macbookpro/Desktop/123.png"
    ]


current_datetime= datetime.now()
datetime_string=current_datetime.strftime("%d%m%Y_%H%M%S")
main_backup_directory = "/Users/macbookpro/Desktop/backup"


backup_zip_file = f"/Users/macbookpro/Desktop/backup_{datetime_string}.zip"


with zipfile.ZipFile (backup_zip_file, "w")as zipf:
    for file_path in backup_list:
        zipf.write(file_path, arcname=file_path.split("/")[-1])



if not os.path.exists(main_backup_directory):
    os.makedirs(main_backup_directory)


```
